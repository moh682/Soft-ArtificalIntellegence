# Mind Your Language

```python
# importing relavent modules
from urllib.request import urlopen;
import re as re;
```

```python
# declaring variables
file_location = './files/txt.txt'
api_url = "http://wdylike.appspot.com"
```

```python
# declaring functions
with open(file_location,'r') as file:     
    sentence = file.read(); # returns the sentence in file
    sentence = sentence.lower() # set all characters to lowercase
    sentence = re.sub('[,.‚ê£]|\s', '', sentence); # remove [ . , ] from sentence and whitespace
    req = urlopen("%s/?q=%s" % (api_url,sentence)) 
    print(req.read().decode('utf-8'))
    
```
