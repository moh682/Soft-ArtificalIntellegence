# Take a Break and Kit Kat

```python
# Importing the proper library
import time as t;
import webbrowser as wb;
from threading import Timer 
```

```python
# setting up variables
time_interval = 5; # in seconds
link_to_youtube = "https://www.youtube.com/watch?v=a6ro6vo1JrQ";
```

```python
# Defining function
def open_song(url):
    wb.get('chrome').open_new(url);
```

```python
# Running the function with a timer 
while True:
    t.sleep(time_interval)
    open_song(link_to_youtube);
```
